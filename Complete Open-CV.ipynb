# Morphological transformations

```python
#Simple operations on image shape(normally performed on binary images)
#2 THINGS REQUIRED: Original image and kernel
#A kernel tells you how to change the value of any given pixel by combining it with different amounts of the neighbouring pixels
img = cv2.imread('D:\Projects\OpenCV\Image\smarties.png', cv2.IMREAD_GRAYSCALE)
_, mask = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)

#To remove black dots
#Dilation
kernal = np.ones((3,3), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations=2)

#Erosion
erosion = cv2.erode(mask, kernal, iterations = 2) 

#Opening-basically erosion followed by dilation
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernal)

#Closing- dilation followed by erosion
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernal)

#Morphological gradient
MG = cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernal)

#Tophat
TH = cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernal)
#There are a lot many 

titles = ['image','mask', 'dilation','erosion', 'opening', 'closing', 'MG','TH']
images = [img,mask,dilation,erosion,opening,closing,MG,TH]

for i in range(8):
    plt.subplot(3,3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
```

# Image Gradients and Edge Detection

```python
img = cv2.imread('D:\Projects\OpenCV\Image\messi5.jpg', cv2.IMREAD_GRAYSCALE)

#Laplacian method
lap = cv2.Laplacian(img, cv2.CV_64F, ksize = 3)
lap = np.uint8(np.absolute(lap))

#Sobel
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize = 1)
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize = 1)

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

#Sobel combined

sobelcombined = cv2.bitwise_or(sobelX,sobelY)

titles = ['image', 'Laplcian method', 'Sobel-x', 'Sobel-y', 'sobelcombined']
images = [img, lap, sobelX, sobelY, sobelcombined]
for i in range(5):
    plt.subplot(2,3,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()
```

# Image blending

```python
orange = cv2.imread('D:\Projects\OpenCV\Image\orange.jpg')
print(orange.shape)
gradient = cv2.imread('D:\Projects\OpenCV\Image\gradient.png')
gradient = cv2.resize(gradient, (512,512))
print(gradient.shape)

gradient_orange = np.hstack((orange[:, :256], gradient[:, :256]))

#Generate Gaussian pyramid
#For orange
orange_copy = orange.copy()
gp_orange = [orange_copy]
for i in range(6):
    orange_copy = cv2.pyrDown(orange_copy)
    gp_orange.append(orange_copy)
    
#For gradient    
gradient_copy = gradient.copy()
gp_gradient = [gradient_copy]
for i in range(6):
    gradient_copy = cv2.pyrDown(gradient_copy)
    gp_gradient.append(gradient_copy)

# Generate Laplacian pyramid
orange_copy = gp_orange[5]
lp_orange = [orange_copy]

for i in range(5, 0, -1):
    gaussian_extended = cv2.pyrUp(gp_orange[i])
    laplacian = cv2.subtract(gp_orange[i-1], gaussian_extended)
    lp_orange.append(laplacian)
    
# Generate Laplacian pyramid
gradient_copy = gp_gradient[5]
lp_gradient = [gradient_copy]

for i in range(5, 0, -1):
    gaussian_extended = cv2.pyrUp(gp_gradient[i])
    laplacian = cv2.subtract(gp_gradient[i-1], gaussian_extended)
    lp_gradient.append(laplacian)
    
#Now add left and right halves of images in each level
orange_gradient_pyramid = []
n =  0 
for orange_lap, gradient_lap in zip(lp_orange, lp_gradient):
    n += 1
    cols, rows, ch = gradient_lap.shape
    laplacian = np.hstack((orange_lap[:, 0:int(cols/2)], gradient_lap[:, int(cols/2):]))
    orange_gradient_pyramid.append(laplacian)
    
#Reconstruct
orange_gradient_reconstruct = orange_gradient_pyramid[0]
for i in range(1, 6):
    orange_gradient_reconstruct = cv2.pyrUp(orange_gradient_reconstruct)
    orange_gradient_reconstruct = cv2.add(orange_gradient_pyramid[i], orange_gradient_reconstruct)

cv2.imshow("orange", orange)
cv2.imshow("gradient", gradient)
cv2.imshow("orange_gradient", gradient_orange)
cv2.imshow("orange_gradient_reconstruct", orange_gradient_reconstruct)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Template Matching

```python
#Template matching is a technique in computer vision used for finding a subimage of a target image which matches a template image.
```

```python
img = cv2.imread("D:\Projects\OpenCV\Image\messi5.jpg")
gray_img = cv2.imread("D:\Projects\OpenCV\Image\messi5.jpg", 0)
#Load Template
template = cv2.imread("D:\Projects\OpenCV\Image\messi5_face.jpg", 0)
w,h = template.shape[::-1]

res = cv2.matchTemplate(gray_img, template, cv2.TM_CCORR_NORMED)
print(res)

threshold = 0.99;
loc = np.where(res >= threshold)
print(loc)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0,0,255),1)

cv2.imshow("Image",img)
cv2.imshow("Template", template)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Hough Line Transform using HoughLines

```python
img = cv2.imread("D:\Projects\OpenCV\Image\sudoku.png")
gray = cv2.imread("D:\Projects\OpenCV\Image\sudoku.png", 0)
edges = cv2.Canny(gray, 50, 150, apertureSize = 3)
lines = cv2.HoughLines(edges, 1, np.pi/180,200)
#Parameters taken:
#Source Image
#rho: Distance resolution of the accumlator in pixels
#theta: Angle resolution of the accumulator in radians
#threshold: Accumulator threshold parameter 
#Lines: Output vector of lines

for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    #x1 stores the rounded off value of (r * cos(theta) - 1000 * sin(theta))
    x1 = int(x0 + 1000 * (-b))
    # y1 stores the rounded off value of (r * sin(theta)+ 1000 * cos(theta))
    y1 = int(y0 + 1000 * (a))
    # x2 stores the rounded off value of (r * cos(theta)+ 1000 * sin(theta))
    x2 = int(x0 - 1000 * (-b))
    # y2 stores the rounded off value of (r * sin(theta)- 1000 * cos(theta))
    y2 = int(y0 - 1000 * (a))
    cv2.line(img, (x1, y1), (x2, y2), (0, 78, 255), 2)
    
cv2.imshow('image', img)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

#  Probabilistic Hough Transform

```python
img = cv2.imread('D:\Projects\OpenCV\Image\sudoku.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
cv2.imshow('edges', edges)
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength=100,maxLineGap=10)
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)

cv2.imshow('image', img)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```
